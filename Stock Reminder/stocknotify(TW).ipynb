```python
import requests
import numpy as np
import time
# 字體顏色
class bcolors:
    OK = '\033[92m' #GREEN
    WARNING = '\033[93m' #YELLOW
    FAIL = '\033[91m' #RED
    RESET = '\033[0m' #RESET COLOR
# LINE Notify 權杖
token = 'F7cbo3hpi6SKYA8KCXXERUSOroaJTZOyO9QaCetMDOU'
# 50日均量計算
def volume_avg(stockID):
    ticker = str(stockID) + '.TW'
    ticker_obj = yf.Ticker(ticker)
    hist = ticker_obj.history(period="50d")
    avg_volume = hist["Volume"].mean()/1000
    if avg_volume != avg_volume:
        ticker = str(ticker) + 'O'
        ticker_obj = yf.Ticker(ticker)
        hist = ticker_obj.history(period="50d")
        avg_volume = hist["Volume"].mean()/1000
    return str(avg_volume)
#預估成交量
def estimate_volume(hour, min, vol):
    hour_range = [9,10,11,12,13]
    min_range = [0,15,30,45]
    estimate_num = [8,5,4,3,2.5,2.2,2,1.8,1.7,1.6,1.5,1.45,1.38,1.32,1.25,1.17,1.1,1]
    if (hour_range[0], min_range[0]) <= (hour, min) < (hour_range[0], min_range[1]):
        vol = vol * estimate_num[0]
    elif (hour_range[0], min_range[1]) <= (hour, min) < (hour_range[0], min_range[2]):
        vol = vol * estimate_num[1]
    elif (hour_range[0], min_range[2]) <= (hour, min) < (hour_range[0], min_range[3]):
        vol = vol * estimate_num[2]
    elif (hour_range[0], min_range[3]) <= (hour, min) < (hour_range[1], min_range[0]):
        vol = vol * estimate_num[3]
    elif (hour_range[1], min_range[0]) <= (hour, min) < (hour_range[1], min_range[1]):
        vol = vol * estimate_num[4]
    elif (hour_range[1], min_range[1]) <= (hour, min) < (hour_range[1], min_range[2]):
        vol = vol * estimate_num[5]
    elif (hour_range[1], min_range[2]) <= (hour, min) < (hour_range[1], min_range[3]):
        vol = vol * estimate_num[6]
    elif (hour_range[1], min_range[3]) <= (hour, min) < (hour_range[2], min_range[0]):
        vol = vol * estimate_num[7]
    elif (hour_range[2], min_range[0]) <= (hour, min) < (hour_range[2], min_range[1]):
        vol = vol * estimate_num[8]
    elif (hour_range[2], min_range[1]) <= (hour, min) < (hour_range[2], min_range[2]):
        vol = vol * estimate_num[9]
    elif (hour_range[2], min_range[2]) <= (hour, min) < (hour_range[2], min_range[3]):
        vol = vol * estimate_num[10]
    elif (hour_range[2], min_range[3]) <= (hour, min) < (hour_range[3], min_range[0]):
        vol = vol * estimate_num[11]
    elif (hour_range[3], min_range[0]) <= (hour, min) < (hour_range[3], min_range[1]):
        vol = vol * estimate_num[12]
    elif (hour_range[3], min_range[1]) <= (hour, min) < (hour_range[3], min_range[2]):
        vol = vol * estimate_num[13]
    elif (hour_range[3], min_range[2]) <= (hour, min) < (hour_range[3], min_range[3]):
        vol = vol * estimate_num[14]
    elif (hour_range[3], min_range[3]) <= (hour, min) < (hour_range[4], min_range[0]):
        vol = vol * estimate_num[15]
    elif (hour_range[4], min_range[0]) <= (hour, min) < (hour_range[4], min_range[1]):
        vol = vol * estimate_num[16]
    elif (hour_range[4], min_range[1]) <= (hour, min) < (hour_range[4], min_range[2]):
        vol = vol * estimate_num[17]
    else:
        vol = vol * estimate_num[17]
    return(str(vol))
# 倉位計算
def position(all_money,confidence,number_of_stocks,price,stop_loss):
    ## 總投資金額
    all_money = all_money
    ## 信心指數
    confidence = confidence
    ## string
    ## 市價
    market_price = price
    ## 停損點
    stop_loss_point = stop_loss
    ## 停損%數
    stop_loss_per = round((market_price-stop_loss_point)/market_price, 4)
    ## 可承擔金額
    afford_loss = all_money*confidence/(number_of_stocks*100)
    ## 投資金額
    invest_money = afford_loss/stop_loss_per
    # print('\n' + f'市價 : {market_price}, 停損點 : {stop_loss_point}, 停損%數 : {stop_loss_per*100}%, \
    # 可承擔損失 : {afford_loss}({round(afford_loss/all_money, 5)*100}%), \
    # 投資金額 : {int(invest_money)}({round(invest_money/all_money, 4)*100}%), {round(int(invest_money)/market_price,1)}股')
    m = [int(invest_money), round(invest_money*100/all_money, 2), int(int(invest_money)/market_price)]
    
    return m
    ##
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time
from random import randint
import pandas as pd
import openpyxl
from tqdm import trange
import yfinance as yf
johnny_allmoney = 230000
jack_allmoney = 100000
# 正常0.5
#曝險大小
risk = 0.8
johnny_confidence = 0.5*risk
jack_confidence = 0.7
johnny_number = 2
jack_number = 4
# 自動下載ChromeDriver
service = ChromeService(executable_path=ChromeDriverManager().install())

# 關閉通知提醒
chrome_options = webdriver.ChromeOptions()

prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)

# 開啟瀏覽器
driver = webdriver.Chrome(service=service, chrome_options=chrome_options)
time.sleep(5)



# 持有股票
stop_loss_stockID = [1108, 6605, 8403]
# 準備買入股票
stockID_list = [3147, 5315]#
# 買入點 # 準備買入股票的停損點
buy_price_list = [71.2, 23.95]# 
newstock_stop_loss = [65.4, 22.6]# 
# 持有股票的停損點
stop_loss_price = [-15.4, -96.4, -34.35]

# 製作全部股票號碼
buy_price_list.extend(stop_loss_price)
price_list = buy_price_list
print(price_list)
stockID_list.extend(stop_loss_stockID)
print(stockID_list)
print(f'price length : {len(price_list)}, ID length : {len(stockID_list)}')
theurl = []
note_text = ['帶大量突破買進',
             '可以直接買，最好帶量',
             '帶大量買進'
]# '可以直接買，突破沒有帶大量部位*0.8(流動性)。','可以直接買，突破沒有帶大量部位*0.8(流動性)。',
# 要發送的訊息
print(f'{bcolors.OK}reset   reset   reset   reset{bcolors.RESET}')
message = '\n【重新設定】 :\n'
for i in range(len(stockID_list)):
    if price_list[i]>0:
        message =  message + str(i+1) + '. ' + str(stockID_list[i]) + '目標股價 : ' + str(price_list[i]) + '\n'
    else:
        message =  message + str(i+1) + '. ' + str(stockID_list[i]) + '停損股價 : ' + str(abs(price_list[i])) + '\n'
# HTTP 標頭參數與資料
headers = { "Authorization": "Bearer " + token }
data = { 'message': message }
connection = 1
# 以 requests 發送 POST 請求
requests.post("https://notify-api.line.me/api/notify",
  headers = headers, data = data)
for i in range(len(stockID_list)):
    # 去到你想要的網頁
    url = 'https://tw.stock.yahoo.com/quote/' + str(stockID_list[i])
    theurl.append(url)
try:
    while(1):
    ########################################################################################改頁數
        for num in range(0,len(theurl)):
            #儲存網址 
            T = np.random.randint(7,size = 1)[0]
            # 去到你想要的網頁
            try:
                driver.get(theurl[num])
                title = driver.find_element(by = By.ID, value = "qsp-overview-realtime-info").text
                #name = driver.find_element(by= By.CLASS_NAME, value = 'D(id)').text
                if connection == 0:
                    requests.post("https://notify-api.line.me/api/notify",
                        headers = { "Authorization": "Bearer " + token },
                        data = { 'message': '\n【Reconnect】' })
                    print(f'{bcolors.OK}Reconnect{bcolors.RESET}')
                    connection = 1
                # time.sleep(T)
            except Exception as e:
                requests.post("https://notify-api.line.me/api/notify",
                    headers = { "Authorization": "Bearer " + token },
                        data = { 'message': '\n【Disconnect】' })
                print(f'{bcolors.WARNING}Go Web Failed{bcolors.RESET}')
                print(e)
                print(f'{bcolors.WARNING}Disconnect{bcolors.RESET}')
                try:
                    driver.close()
                except Exception as e:
                    print(e)
                    print(f'{bcolors.WARNING}Close Google Failed{bcolors.RESET}')
                    pass
                time.sleep(2)
                # service = ChromeService(executable_path=ChromeDriverManager().install())
                driver = webdriver.Chrome(service=service, chrome_options=chrome_options)
                connection = 0
                continue
            time.sleep((3))
            try:
                title = title.split('\n')
                name = title[0]
                data_time = title[1].split('：')[1]
                print(stockID_list[num])
                print(data_time)
                hour = data_time.split(' ')[1].split(':')[0]
                min = data_time.split(' ')[1].split(':')[1]
                price = title[4]
                now_volume = title[22]
                yesterday_volume = title[24]
                volume50avg = volume_avg(str(stockID_list[num]))
                #print(now_volume.replace(',',''))
                est_volume = estimate_volume(int(hour), int(min), int(now_volume.replace(',','')))
            except Exception as e:
                print(f'{bcolors.FAIL}Get Data Failed{bcolors.RESET}')
                print(e)
            try:
                if (float(price) > abs(price_list[num])) & (price_list[num]>0):
                    # 要發送的訊息
                    message = f'\n【{str(stockID_list[num])}】\n ⚠️ : {note_text[num]}\n{data_time}\n目標股價 : {str(price_list[num])}\n\
    現價 : {price}(超過{str(round(100*(float(price) - price_list[num])/price_list[num],2))}%)\
        \n今日總量 : {now_volume}\n今日預估量 : {est_volume}\n50日均量 : {int(float(volume50avg))}\n昨日總量 : {yesterday_volume}'
                    if stockID_list[num] not in stop_loss_stockID:
                        if float(est_volume) > float(volume50avg):
                            johnny_position = position(johnny_allmoney,johnny_confidence,
                                                    johnny_number,float(price),newstock_stop_loss[num])
                            jack_position = position(jack_allmoney,jack_confidence,
                                                    jack_number,float(price),newstock_stop_loss[num])
                            if johnny_position[0] >= johnny_allmoney*0.06:
                                johnny_position = [johnny_allmoney*0.06, 6, johnny_allmoney*0.06/float(price)]
                            johnny_position_text = f'投資金額 : {int(johnny_position[0])}({johnny_position[1]}%), \
        {int(johnny_position[2])}股'
                            jack_position_text = f'投資金額 : {int(jack_position[0])}({jack_position[1]}%), \
        {int(jack_position[2])}股'
                            message = message + f'\n停損股價 : {newstock_stop_loss[num]}\nJohnny 2️⃣: {johnny_position_text}\nJack : {jack_position_text}'
                        else :
                            johnny_position = position(johnny_allmoney,johnny_confidence,
                                                    johnny_number,float(price),newstock_stop_loss[num])
                            jack_position = position(jack_allmoney,jack_confidence,
                                                    jack_number,float(price),newstock_stop_loss[num])
                            if johnny_position[0] >= johnny_allmoney*0.06:
                                johnny_position = [johnny_allmoney*0.06, 6, johnny_allmoney*0.06/float(price)]
                            johnny_position_text = f'投資金額 : {int(johnny_position[0])}({johnny_position[1]}%), \
        {int(johnny_position[2])}股'
                            jack_position_text = f'投資金額 : {int(jack_position[0])}({jack_position[1]}%), \
        {int(jack_position[2])}股'
                            
                            message = message + f'\n停損股價 : {newstock_stop_loss[num]}\nJohnny 1️⃣ : {johnny_position_text}\nJack : {jack_position_text}'
                    print(message)
                    # HTTP 標頭參數與資料
                    headers = { "Authorization": "Bearer " + token }
                    data = { 'message': message }

                    # 以 requests 發送 POST 請求
                    requests.post("https://notify-api.line.me/api/notify",
                        headers = headers, data = data)
                elif (float(price) < abs(price_list[num])) & (price_list[num]<0):
                    # 要發送的訊息
                    message = f'\n【{str(stockID_list[num])}】\n停損股價 : {str(abs(price_list[num]))}\n現價 : {price}({str(100*round((float(price) - abs(price_list[num]))/abs(price_list[num]),4))}%)'
                    print(message)
                    # HTTP 標頭參數與資料
                    headers = { "Authorization": "Bearer " + token }
                    data = { 'message': message }

                    # 以 requests 發送 POST 請求
                    requests.post("https://notify-api.line.me/api/notify",
                        headers = headers, data = data)
                print('---------------------------------------------------------------------------------------------------------------------------------------------')
            except Exception as e:
                print(e)
                print(f'{bcolors.FAIL}POST TEXT FAILED{bcolors.RESET}')
except Exception as e:
    print(e)
    print(f'{bcolors.FAIL}Program Shutdown{bcolors.RESET}')
    message = '\n【Program Shutdown】'
    headers = { "Authorization": "Bearer " + token }
    data = { 'message': message }
    requests.post("https://notify-api.line.me/api/notify",
                  headers = headers, data = data)
    # time.sleep(60)
        
```
