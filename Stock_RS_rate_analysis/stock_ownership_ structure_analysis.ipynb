```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
stock_nums = '2493 3260 5439 3530 3293 3615 2402 3596'.split(' ')
stock_nums = '3596'.split(' ')
path = 'C:/Users/User/Desktop/StockInfoHub/Stock_Data_Collector/股東人數資料/'
plt.rcParams['font.size'] = 15
for stock_num in stock_nums:
    try:
        df = pd.read_excel(f'{path}{stock_num}.TW.xlsx')
    except:
        df = pd.read_excel(f'{path}{stock_num}.TWO.xlsx')
    # avg_share = df['平均張數/人'].astype(float)
    num_of_share_holder = df['總股東人數'].map(lambda x: int(x.replace(',', ''))).astype(int)
    more_than_1000_share = df['>1000張大股東持有百分比'].astype(int)
    more_than_400_share = df['>400張大股東持有百分比'].astype(int)
    date = df['資料日期']
    close_price = df['收盤價'].astype(float)
    #plot bar chart with left y axis, plot line chart with right y axis
    # plt.figure(figsize=(20, 10))
    fig, ax1 = plt.subplots()
    fig.set_size_inches(30, 15)
    ax2 = ax1.twinx()
    ax3 = ax1.twinx()
    ax3.plot(close_price.values[::-1], color='black', label='close price', linewidth=3)
    ax3.set_axis_off()
    ax4 = ax1.twinx()
    ax1.bar(num_of_share_holder.index, num_of_share_holder.values[::-1], color='g', alpha=0.5, width=0.8, align='center', label='number of share holder')
    # ax2.plot(avg_share.index, avg_share.values[::-1], color='r', label='avg share per person')
    ax2.plot(more_than_1000_share.index, more_than_1000_share.values[::-1], color='r', label='Percentage of Ownership Held by Shareholders with >1000 Shares')
    ax2.plot(more_than_400_share.index, more_than_400_share.values[::-1], color='b', label='Percentage of Ownership Held by Shareholders with >400 Shares')
    ax1.set_xlabel('date', fontsize=18)
    ax1.set_ylabel('number of share holder', fontsize=18)
    ax2.set_ylabel('Percentage of Ownership(%)', fontsize=18)
    
    ax1.set_ylim(min(num_of_share_holder.values[::-1]) * 0.5, max(num_of_share_holder.values[::-1]) * 1.5)
    ax2.set_ylim(min(more_than_400_share.values[::-1]) * 0.5, max(more_than_400_share.values[::-1]) * 1.5)
    
    ax1.legend(loc='upper left', fontsize=18)
    ax2.legend(loc='upper right', fontsize=18)
    ax3.legend(loc='lower right', fontsize=18)
    ax1.grid(axis='y')
    ax2.set_xticks(np.arange(0, len(date), 10))
    ax2.set_xticklabels(date[np.arange(0, len(date), 10)][::-1], rotation=30, fontsize=15)
    plt.title(f'{stock_num}')
    plt.show()
```
