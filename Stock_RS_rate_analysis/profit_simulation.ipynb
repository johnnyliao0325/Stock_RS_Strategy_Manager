```python
import numpy as np
import matplotlib.pyplot as plt

mu = -0.6  # 平均值
sigma = 8  # 标准差
data = np.random.normal(mu, sigma, 5000)  # 生成600個正常獲利
data2 = np.random.normal(mu-1, sigma, 5000)
data = data[data<20]
data = data[data>-12]
data2 = data2[data2<20]
data2 = data2[data2>-10]
# print(len(data))
# cut = data[(((data <= 0) & (data>-2.5)))]
# print(np.mean(cut))
# plt.hist(cut, bins = 10)
# plt.show()
# data = data[((data >= 0) | (data<-2.5))]
# 绘制直方图
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
plt.hist(data2, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
print(np.mean(data))
print(len(data[data>0]))
print(len(data[data<=0]))
print(np.mean(data2))
print(len(data2[data2>0]))
print(len(data2[data2<=0]))
```

```python
import random

def random_win():
    choices = [0] * 14 + [1] * 86  # 构造长度为100的选择列表
    result = random.choices(choices)  # 随机选择一个数
    return result[0]  # 返回随机选择的数
def random_lose():
    choices = [0] * 2 + [1] * 98  # 构造长度为100的选择列表
    result = random.choices(choices)  # 随机选择一个数
    return result[0]  # 返回随机选择的数
```

```python
year = 130
t = 2000
```

```python
# 沒有調整倉位
allprofit_list = []
position = 0.16
profit_list = []
avgp = []
addtimes_list = []
win_lsit = []
lose_list = []
max_list = []
min_list = []
nt = 0
d1 = data2
fsize = 20
fsize2 = 12
for j in range(t):
    allprofit = 0
    minprofit_times = 0
    maxprofit_times = 0
    win_times = 0
    lose_times = 0
    p = []
    addtimes=0
    lastprofit = 0
    neg = 0
    allmoney = 600000
    for i in range(year):
        stop_loss = random.choice(d1[d1<=-2.5])
        profit_percentage = random.choice(d1)
        if profit_percentage <= 0:
            profit_percentage = random.choice(d1[d1<=0])*random_lose()
            if not profit_percentage:
                profit_percentage = random.choice(mindata)
                minprofit_times += 1
            lose_times += 1
        else:
            profit_percentage = profit_percentage*random_win()
            if not profit_percentage:
                profit_percentage = random.choice(maxdata)
                maxprofit_times += 1
            win_times += 1
        p.append(profit_percentage/100)
        profit = profit_percentage*allmoney*position/100
        allprofit_list.append(profit)
        allprofit += profit
        # 複利
        if ((allprofit-lastprofit)/allmoney) > 0.2:
            allmoney = allmoney+allprofit
            lastprofit = allprofit
            addtimes += 1
        if profit<0 :
            neg += 1
        if profit>0:
            neg = 0
        if neg >= 10:
            nt += 1
    # print(f'獲勝次數:{win_times}, 失敗次數:{lose_times}, 爆贏次數:{maxprofit_times}, 爆輸次數:{minprofit_times}')
    # print(f'總獲利:{allprofit}, 平均%數:{np.mean(p)}')
    # 每次平均獲利%
    avgp.append(np.mean(p))
    # 每次總獲利金額
    profit_list.append(allprofit)
    addtimes_list.append(addtimes)
    win_lsit.append(win_times)
    lose_list.append(lose_times)
    max_list.append(maxprofit_times)
    min_list.append(minprofit_times)
print(f'全部平均獲利:{np.mean(profit_list)}({round(np.mean(profit_list)/600000,4)*100}%), 平均複利次數:{np.mean(addtimes_list)}, 獲利cv:{np.std(profit_list)}, 平均獲利%:{np.mean(avgp)}, 獲利%數cv:{np.std(avgp)/np.mean(avgp)}')
print(f'最小/大獲利:{min(profit_list)}/{max(profit_list)}, 平均獲勝次數:{np.mean(win_lsit)}, 平均失敗次數:{np.mean(lose_list)}, 平均爆贏次數:{np.mean(max_list)}, 平均爆輸次數:{np.mean(min_list)}')
print(nt)
plt.figure(figsize = (22,11))
plt.hist(allprofit_list, bins=100)
plt.grid()
plt.yticks(np.arange(0,400,50),fontsize=fsize2)
plt.xticks(np.arange(-50000,150000,10000),fontsize=fsize2)
plt.ylabel('times',fontsize=fsize)
plt.xlabel('profit',fontsize=fsize)
plt.show()
print(np.std(np.array(allprofit_list)[np.array(allprofit_list)<0]))
allprofit_list = np.cumsum(allprofit_list)
```
