```python
import matplotlib.pyplot as plt

name_list = ['Long', 'Short', 'Rebound', 'Pullback', 'consolidation']
data_list = [long, short, rebound, pullback, consolidation]

# Strategies' profit columns
profit_columns = ['T5 S9 profit', 'T5 S7 profit', 'T6 S9 profit', 'T6 S7 profit'
                  , 'T11 S9 profit', 'T11 S7 profit', 'TM S9 profit', 'TM S7 profit']
# profit_columns = ['T5 S4 profit', 'T6 S4 profit', 'T11 S4 profit', 'TM S4 profit']
# profit_columns2 = ['T5 NS profit', 'T6 NS profit', 'T11 NS profit', 'TM NS profit']
# all_column = np.concatenate([profit_columns,profit_columns2]).reshape((8,1))
print(profit_columns)
# Plot boxplots for each category and strategy
plt.rcParams['axes.facecolor'] = '#F5F5F5'  # 将背景色设为浅灰色
plt.rcParams['figure.facecolor'] = '#F5F5F5'  # 将图形的背景色设为浅灰色
plt.rcParams['font.size'] = 20  # 设置字体大小为12
for i in range(5):
    print(name_list[i])
    fig, ax = plt.subplots(figsize=(26, 15))  # 調整 figsize 的大小
    bp = ax.boxplot([data_list[i][col] for col in profit_columns], vert=False)
    # bp2 = ax.boxplot([data_list[i][col] for col in profit_columns2], vert=False)
    ax.set_title(name_list[i], fontsize=20)  # 調整標題的大小
    ax.set_yticklabels(profit_columns, fontsize=16)  # 調整 y 軸刻度標籤的大小
    ax.set_xlabel('Profit %', fontsize=20)  # 調整 x 軸標籤的大小
    ax.set_xticklabels(ax.get_xticks(), fontsize=20)
    ax.xaxis.grid(True)
    for profit in profit_columns:
        # Add important values for T6 2M profit
        t6_values = data_list[i][profit]
        t6_q1 = t6_values.quantile(0.25)
        t6_q3 = t6_values.quantile(0.75)
        t6_median = t6_values.median()
        t6_min = t6_values.min()
        t6_max = t6_values.max()
        t6_mean = t6_values.mean()
        t6_win = (t6_values>0.3).sum()
        
        for j, box in enumerate(bp['boxes']):
            if profit_columns[j] == profit:
                box_x = box.get_xdata().mean()
                box_y = box.get_ydata().max() + 0.1
                ax.text(box_x, box_y, f"Min: {t6_min:.2f} | Q1: {t6_q1:.2f} | Median: {t6_median:.2f} | Mean: {t6_mean:.2f} | Q3: {t6_q3:.2f} | Max: {t6_max:.2f} | Win rate: {round(100*t6_win/len(t6_values),2)}%",
                        verticalalignment='bottom', horizontalalignment='center', fontsize=16)

    plt.tight_layout()
    path = 'C:/Users/User/Desktop/code/python/stockdata_VCP/VCP_predict_project/ETF test/多頭策略統計比較/'+'停利停損策略S4S5'+name_list[i] 
    # plt.savefig(path, bbox_inches='tight')
    plt.show()
    
    # display(data_list[i].describe())
```
