```python
import numpy as np
import pandas as pd
data = pd.read_excel(r'C:\Users\User\Desktop\code\python\stockdata_VCP\VCP_predict_project\ETF test\all short template (1Week).xlsx')
data.describe()
```

```python
long = data.loc[data['週期']=='多']
short = data.loc[data['週期']=='空']
rebound = data.loc[data['週期']=='反彈']
pullback = data.loc[data['週期']=='拉回']
consolidation = data.loc[data['週期']=='盤整']
name_list = ['多','空','反彈','拉回','盤整']
data_list = [long, short, rebound, pullback, consolidation]
```

```python
import matplotlib.pyplot as plt

name_list = ['Long', 'Short', 'Rebound', 'Pullback', 'consolidation']
data_list = [long, short, rebound, pullback, consolidation]

# Strategies' profit columns
profit_columns = ['ST5 1M profit', 'ST6 1M profit', 'TR1 1M profit', 'TR2 1M profit']
# Plot boxplots for each category and strategy
plt.rcParams['axes.facecolor'] = '#F5F5F5'  # 将背景色设为浅灰色
plt.rcParams['figure.facecolor'] = '#F5F5F5'  # 将图形的背景色设为浅灰色
plt.rcParams['font.size'] = 20  # 设置字体大小为12
for i in range(5):
    fig, ax = plt.subplots(figsize=(16, 10))  # 調整 figsize 的大小
    bp = ax.boxplot([data_list[i][col] for col in profit_columns], vert=False)
    ax.set_title(name_list[i], fontsize=20)  # 調整標題的大小
    ax.set_yticklabels(profit_columns, fontsize=16)  # 調整 y 軸刻度標籤的大小
    ax.set_xlabel('Profit %', fontsize=20)  # 調整 x 軸標籤的大小
    ax.set_xticklabels(ax.get_xticks(), fontsize=20)
    ax.xaxis.grid(True)

    # Add important values for T6 1M profit
    t6_values = data_list[i]['ST5 1M profit']
    t6_q1 = t6_values.quantile(0.25)
    t6_q3 = t6_values.quantile(0.75)
    t6_median = t6_values.median()
    t6_min = t6_values.min()
    t6_max = t6_values.max()
    
    for j, box in enumerate(bp['boxes']):
        if profit_columns[j] == 'ST5 1M profit':
            box_x = box.get_xdata().mean()
            box_y = box.get_ydata().max() + 0.1
            ax.text(box_x, box_y, f"Min: {t6_min:.2f} | Q1: {t6_q1:.2f} | Median: {t6_median:.2f} | Q3: {t6_q3:.2f} | Max: {t6_max:.2f}",
                    verticalalignment='bottom', horizontalalignment='center', fontsize=16)
    # Add important values for T6 1M profit
    t6_values = data_list[i]['ST6 1M profit']
    t6_q1 = t6_values.quantile(0.25)
    t6_q3 = t6_values.quantile(0.75)
    t6_median = t6_values.median()
    t6_min = t6_values.min()
    t6_max = t6_values.max()
    
    for j, box in enumerate(bp['boxes']):
        if profit_columns[j] == 'ST6 1M profit':
            box_x = box.get_xdata().mean()
            box_y = box.get_ydata().max() + 0.1
            ax.text(box_x, box_y, f"Min: {t6_min:.2f} | Q1: {t6_q1:.2f} | Median: {t6_median:.2f} | Q3: {t6_q3:.2f} | Max: {t6_max:.2f}",
                    verticalalignment='bottom', horizontalalignment='center', fontsize=16)
    # Add important values for T6 1M profit
    t6_values = data_list[i]['TR1 1M profit']
    t6_q1 = t6_values.quantile(0.25)
    t6_q3 = t6_values.quantile(0.75)
    t6_median = t6_values.median()
    t6_min = t6_values.min()
    t6_max = t6_values.max()
    
    for j, box in enumerate(bp['boxes']):
        if profit_columns[j] == 'TR1 1M profit':
            box_x = box.get_xdata().mean()
            box_y = box.get_ydata().max() + 0.1
            ax.text(box_x, box_y, f"Min: {t6_min:.2f} | Q1: {t6_q1:.2f} | Median: {t6_median:.2f} | Q3: {t6_q3:.2f} | Max: {t6_max:.2f}",
                    verticalalignment='bottom', horizontalalignment='center', fontsize=16)
    # Add important values for T6 1M profit
    t6_values = data_list[i]['TR2 1M profit']
    t6_q1 = t6_values.quantile(0.25)
    t6_q3 = t6_values.quantile(0.75)
    t6_median = t6_values.median()
    t6_min = t6_values.min()
    t6_max = t6_values.max()
    
    for j, box in enumerate(bp['boxes']):
        if profit_columns[j] == 'TR2 1M profit':
            box_x = box.get_xdata().mean()
            box_y = box.get_ydata().max() + 0.1
            ax.text(box_x, box_y, f"Min: {t6_min:.2f} | Q1: {t6_q1:.2f} | Median: {t6_median:.2f} | Q3: {t6_q3:.2f} | Max: {t6_max:.2f}",
                    verticalalignment='bottom', horizontalalignment='center', fontsize=16)

    plt.tight_layout()
    path = 'C:/Users/User/Desktop/code/python/stockdata_VCP/VCP_predict_project/ETF test/空頭策略統計比較/'+name_list[i]
    plt.savefig(path,bbox_inches='tight')
    # plt.show()
    print(name_list[i])
    display(data_list[i].describe())
```
